```python
# r+ read + Write
# creates if not present and opens for writing

with open('counter.txt','w') as f:
    f.write('Counter: 0')
f.close()


with open('counter.txt', 'r+') as f:
    content=f.read()
    print("Before data:\n",content)

    f.seek(0)
    f.write('Counter: 1')
```
